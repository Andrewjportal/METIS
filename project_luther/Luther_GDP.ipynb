```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import requests
```

```python
gdp = pd.read_html("https://www.cia.gov/library/publications/the-world-factbook/fields/208.html#AO")
```

```python
df_gdp = pd.DataFrame(gdp[0])
```

```python
df_gdp['GDP_1'] = df_gdp['GDP (purchasing power parity)']
```

```python
df_gdp['GDP_1'] = df_gdp['GDP (purchasing power parity)'].str.split('(', n = 1, expand = True)
```

```python
df_gdp.head(50)
```

```python
df_gdp['GDP_1'] = df_gdp['GDP_1'].replace({'[A-Za-z]':''}, regex=True)
```

```python
df_gdp['GDP_1'] = df_gdp['GDP_1'].replace({'[$]':''}, regex=True)
```

```python
df_gdp['GDP_1'] = pd.to_numeric(df_gdp['GDP_1'], errors='coerce')
```

```python
df_gdp['mil']=df_gdp['GDP (purchasing power parity)'].str.contains("million")
```

```python
df_gdp['bil']=df_gdp['GDP (purchasing power parity)'].str.contains("billion")
```

```python
df_gdp['tril']=df_gdp['GDP (purchasing power parity)'].str.contains("trillion")
```

```python
df_gdp.loc[df_gdp['mil'].eq(True) & df_gdp['GDP_1'],'GDP_1'] = df_gdp['GDP_1'].multiply(1000000)
```

```python
df_gdp.loc[df_gdp['bil'].eq(True) & df_gdp['GDP_1'],'GDP_1'] = df_gdp['GDP_1'].multiply(1000000000)
```

```python
df_gdp.loc[df_gdp['tril'].eq(True) & df_gdp['GDP_1'],'GDP_1'] = df_gdp['GDP_1'].multiply(1000000000000)
```

```python
df_gdp=df_gdp.drop(df_gdp.columns[[1,3,4,5]], axis=1)
```
